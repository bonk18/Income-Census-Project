## Data preprocessing :
1. outliers removal- IQR
2. data normalisation + power transformer (for skewness)
3. vif value for multicollinearity

```python
q1=data.quantile(0.25)
q3=data.quantile(0.75)
IQR=q3-q1
IQR
```

Outliers detection formula:
<br>higherside: Q3+(1.5&#42;IQR)
<br>lowerside: Q3-(1.5&#42;IQR)

```python
itv_high=(q3.fnlwgt+IQR.fnlwgt)
print(itv_high)
index=np.where(data['fnlwgt']>itv_high)
data=data.drop(data.index[index])
print(data.shape)
data.reset_index()
```

```python
itv_high=(q3.age+IQR.age)
print(itv_high)
index=np.where(data['age']>itv_high)
data=data.drop(data.index[index])
print(data.shape)
data.reset_index()
```

```python
plt.figure(figsize=(5,5))
sns.boxplot(data=data['fnlwgt'],orient='v')
plt.xlabel('Final weight',fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.boxplot(data=data['age'],orient='v')
plt.xlabel('Age',fontsize=12)
plt.show()
```

```python
features=['age','fnlwgt']
from sklearn.preprocessing import PowerTransformer
pt=PowerTransformer(method='box-cox')
data[features]=pt.fit_transform(data[features].values)
data.head()
```

### Note:
power transformer has two modes:
1. Box-Cox: used when the features are positive
2. Yeo-Johnson: used when the features are both positive and negative

### Model Training 

```python
from sklearn.linear_model import LogisticRegression
LR=LogisticRegression()
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
for i in range(500,700):
    x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.20,random_state=i)
    LR.fit(x_train,y_train)
    pred_train=LR.predict(x_train)
    pred_test=LR.predict(x_test)
    if round(accuracy_score(y_train,pred_train)*100,1)==round(accuracy_score(y_test,pred_test)*100,1):
        print(f'At random state {i} model performs very well ')
        print(f'At random state {i}')
        print(f'Training accuracy score is {accuracy_score(y_train,pred_train)*100}')
        print(f'Testing accuracy score is {accuracy_score(y_test,pred_test)*100}')
```

### Random_State = 631

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.20,random_state=631)
```

```python
from imblearn.over_sampling import SMOTE
from collections import Counter
```

```python
print('Training outcome \n',y_train.value_counts())
```

```python
Counter(y_train)
```

```python
ove_smp=SMOTE(sampling_strategy=0.80)
x_train_ns,y_train_ns=ove_smp.fit_resample(x_train,y_train)
print('the number of classes before fit',format(Counter(y_train)))
print('the number of classes after fit',format(Counter(y_train_ns)))
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,confusion_matrix
lr=LogisticRegression()
lr.fit(x_train_ns,y_train_ns)
pred=lr.predict(x_test)
print(f'Accuracy score {accuracy_score(y_test,pred)*100}')
print(f'Classification report \n {classification_report(y_test,pred)}')
print(f'Confusion matrix  \n{confusion_matrix(y_test,pred)}')
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(x_train_ns,y_train_ns)
pred_dt=dt.predict(x_test)
print(f'Accuracy score {accuracy_score(y_test,pred_dt)*100}')
print(f'Classification report \n {classification_report(y_test,pred_dt)}')
print(f'Confusion matrix  \n{confusion_matrix(y_test,pred_dt)}')
```

### Ensemble Technique

#### Random Forset

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train_ns,y_train_ns)
pred_rf=rf.predict(x_test)
print(f'Accuracy score {accuracy_score(y_test,pred_rf)*100}')
print(f'Classification report \n {classification_report(y_test,pred_rf)}')
print(f'Confusion matrix  \n{confusion_matrix(y_test,pred_rf)}')
```

#### Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
gb=GradientBoostingClassifier()
gb.fit(x_train_ns,y_train_ns)
pred_gb=gb.predict(x_test)
print(f'Accuracy score {accuracy_score(y_test,pred_gb)*100}')
print(f'Classification report \n {classification_report(y_test,pred_gb)}')
print(f'Confusion matrix  \n{confusion_matrix(y_test,pred_gb)}')
```

#### Xtreme Boosting

```python
from xgboost import XGBClassifier
xgb=XGBClassifier()
xgb.fit(x_train_ns,y_train_ns)
pred_xgb=xgb.predict(x_test)
print(f'Accuracy score {accuracy_score(y_test,pred_xgb)*100}')
print(f'Classification report \n {classification_report(y_test,pred_xgb)}')
print(f'Confusion matrix  \n{confusion_matrix(y_test,pred_xgb)}')
```

```python
from sklearn.svm import SVC
svc=SVC()
svc.fit(x_train_ns,y_train_ns)
pred_svc=svc.predict(x_test)
print(f'Accuracy score {accuracy_score(y_test,pred_svc)*100}')
print(f'Classification report \n {classification_report(y_test,pred_svc)}')
print(f'Confusion matrix  \n{confusion_matrix(y_test,pred_svc)}')
```

### Hyper parameter tuning for Xtreme boosting
